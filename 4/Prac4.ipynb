```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.array([95,85,80,70,60])
y=np.array([85,95,70,65,70])
```

```python
model=np.polyfit(x,y,1)
model
```

```python
predict=np.poly1d(model)
predict(65)
```

```python
y_pred=predict(x)
print(y_pred)
```

```python
from sklearn.metrics import r2_score
r2=r2_score(y,y_pred)
print("R-squared:",r2)
```

```python
y_line = model[1] + model[0] * x
plt.plot(x, y_line, c='r', label='Fitted Line')
plt.scatter(x, y, c='b', label='Data Points')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Liner Regression Model')
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
sp = pd.read_csv(r"C:\Users\Lin-X\Desktop\DSBDA PRAC/Boston.csv")
```

```python
x = sp.drop(['crim'], axis=1)
y = sp['crim']
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2)
```

```python
lm = LinearRegression()
model = lm.fit(xtrain, ytrain)
```

```python
ytrain_pred = lm.predict(xtrain)
ytest_pred = lm.predict(xtest)
```

```python
plt.scatter(ytrain, ytrain_pred, c='blue', marker='o', label='Training Data')
plt.scatter(ytest, ytest_pred, c='red', marker='p', label='Test data')
plt.xlabel('True values')
plt.ylabel('Predicted ')
plt.title("True Value vs Predicted value")
plt.legend(loc='lower right')
plt.show()
```
